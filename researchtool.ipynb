****CharacterTextSplitter****

```python
from langchain.text_splitter import CharacterTextSplitter

# Create an instance of the splitter
splitter = CharacterTextSplitter(
    separator="\n",  # Split text at newline characters
    chunk_size= 200,  # Create chunks of up to 200 characters
    chunk_overlap=0  # No overlap between chunks
)

# Assuming you have the text to split in a variable named `text`
chunks = splitter.split_text(text)  # Call the split_text method on the instance

# Now you can use the `chunks` list, which contains the split text segments
print(chunks)  # Example usage to print the chunks
```

```python
len(chunks)
```

```python
for chunk in chunks :
    print(len(chunk))
```

**RecursiveCharacterTextSplitter**

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
r_splitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n" , "\n", " ", "."],
    chunk_size = 200,
    chunk_overlap = 0
)
chunks = r_splitter.split_text(text)
```

**HERE WE ARE CONVERTING THE WORD COLUMN INTO EMBEDDINGS**

```python
#pip install faiss-cpu
#pip install sentence-transformers
import pandas as pd
pd.set_option('display.max_colwidth', 100)
```

```python
df = pd.read_csv('sample_text.csv')
df
```

```python
#sentence_transformers is a package from hugging face to create embeddings
#Purpose: It maps sentences and short paragraphs into a 768-dimensional dense vector space, capturing their semantic meaning. This makes it ideal for tasks like:

from sentence_transformers import SentenceTransformer
encoder = SentenceTransformer("all-mpnet-base-v2")
vectors = encoder.encode(df['text'])

vectors
```

```python
dim = vectors.shape[1]
dim
```

**we will store the embeddings into a FAISS INDEX(vectpr database)**

```python
import faiss
index = faiss.IndexFlatL2(dim)    #This creates a Faiss index that uses Euclidean distance for similarity calculations.The dim argument specifies the dimensionality of the vectors to be indexed.
index
```

```python
index.add(vectors)
```

**search an embedding in vectordatabase**

```python
search_query = "lets go for a holiday tour"

#convert these search query into vectors
vec = encoder.encode(search_query)
vec.shape
```

```python
import numpy as np 
svec = np.array(vec).reshape(1,-1)
svec
```

```python
 #this index.search() this expects 2d array
distances , I = index.search(svec , k = 2)
I
```

```python
df.loc[I[0]]
```

(1) Load data

```python
#pip install python-libmagic
loader = UnstructuredURLLoader(
    urls=["https://www.moneycontrol.com/competition/tatamotors/consolidated-comparison/TM03",
          "https://www.tata.com/investors/stock-data"])
data = loader.load() 
len(data)
```

(2) Split data to create chunks

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

# As data is of type documents we can directly use split_documents over split_text in order to get the chunks.
docs = text_splitter.split_documents(data)
len(docs)
```

```python
docs[9]
```

(3) Create embeddings for these chunks and save them to FAISS index

```python
# Create the embeddings of the chunks using openAIEmbeddings
embeddings = OpenAIEmbeddings()

# Pass the documents and embeddings inorder to create FAISS vector index
vectorindex_openai = FAISS.from_documents(docs, embeddings)
```

```python
# Storing vector index create in local
import pickle
file_path="vector_index.pkl"
with open(file_path, "wb") as f:
    pickle.dump(vectorindex_openai, f)
```
